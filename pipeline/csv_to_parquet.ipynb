```python
# import libraries
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df_pp = pd.read_csv('data/bats_primary_productivity.csv')
df_bot = pd.read_csv('data/bats_bottle.csv')
df_ctd = pd.read_csv('data/bats_ctd.csv')
df_pig = pd.read_csv('data/bats_pigments.csv')

# Write the DataFrame to a Parquet file
df_pp.to_parquet('bats_pp.parquet', engine='fastparquet', index=False)
df_bot.to_parquet('bats_bot.parquet', engine='fastparquet', index=False)
df_ctd.to_parquet('bats_ctd.parquet', engine='fastparquet', index=False)
df_pig.to_parquet('bats_pig.parquet', engine='fastparquet', index=False)
print(f"Successfully converted csv files to parquet.")
```
